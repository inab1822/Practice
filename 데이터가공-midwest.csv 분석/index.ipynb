#### 2. 변수(컬럼)명 변경 수정
  - poptotal(전체 인구) -> total
  - popasain(아시아 인구) -> asian

```python
mw = mw.rename(columns={'poptotal':'total',
                'popasian':'asain'})
```

```python
mw.columns
```

#### 3. 파생변수 만들어주세요. 
  - total, asian 변수 활용해서 '전체 인구대비 아시아인구 백분율' 
  - 히스토그램으로 분포 살펴보세요.


```python
mw = mw.assign(asian_per_total=lambda x : (x['asain']/x['total'])*100)
mw
```

```python
mw['asian_per_total'].plot.hist()
```

#### 4. '전체 인구대비 아시아인구 백분율' 평균을 구하고, 평균 초과는 'large' 아니면 'small' 파생변수를 만들어주세요.

```python
import numpy as np
mw = mw.assign(size= np.where(mw['asian_per_total']> mw['asian_per_total'].mean(),
                         'large','small'))
mw
```

#### 5. 4번에서 만든 파생변수 지역이 얼마나 많은지 막대그래프를 만들어주세요.

```python
mw['size'].value_counts()
```

```python
mw['size'].value_counts().plot.bar(rot=0)
# rot = 0 을하면 x축의 small과 large 가 가로정렬된다.
```

#### 6. 'popadults'는 성인 인구, 'total' 전체인구. '전체인구 대비 미성년 인구 백분율' 변수를 추가해주세요.

```python
mw = mw.assign(
    미성년인구백분율 = lambda x : ((x['total']-x['popadults'])/x['total']) * 100
)
```

```python
mw
```

#### 7. '전체인구 대비 미성년 인구 백분율'이 가장 높은 상위 5개 지역의 미성년인구백분율을 출력해주세요.

```python
mw.sort_values('미성년인구백분율',ascending=0)[['county','미성년인구백분율']]\
    .head(5)
```

#### 8. 미성년비율등급 변수 추가해주세요.
  - large : 40%이상
  - middle : 30~40%
  - small : 30%미만

```python
mw = mw.assign(
    미성년인구등급 = np.where(mw['미성년인구백분율']>= 40,'large',
                       np.where(mw['미성년인구백분율']<30,'small','middle'))
)
```

```python
mw
```

```python
mw['미성년인구등급'].value_counts()
```
